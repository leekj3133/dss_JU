시간데이터를  정의에 따라 데이터 문자열로 형변환

#새벽 2 ~ 6 , 6 ~ 11, 11 ~ 15,점심 ,15 ~ 18오후,  18 ~ 22 저녁, 22 ~ 02심야


```python
#우선 시간데이터의 비교를 편하기 하기위해 00,01,02시를 24시,25시,26시으로 바꿔주고 시간데이터를 정수 데이터로 바꿔준다
new_filltered_df["time"] = new_filltered_df["time"].replace(["00시","01시","02시"],["24시","25시","26시"])
new_filltered_df["time"].unique()
```

```python
#조건식을 만들어 시간 데이터를 정의에 맞게 다시 문자열 데이터로 형변환 하세요
# 2~6 새벽, 7~11 아침, 12 ~ 15,점심 ,16~18오후, 19~22 저녁, 23~26심야
time = new_filltered_df["time"].unique()
time = new_filltered_df["time"].convert_dtypes()
time = time.str.strip("시")
time = time.astype(int)
time
```

```python
def test(times):
    if 2<=times<=6:
        time_str = "새벽"
    elif 7<=times<=11:
        time_str = "아침"
    elif 12<=times<=15:
        time_str = "점심"
    elif 16<=times<=18:
        time_str = "오후"
    elif 19<=times<=22:
        time_str = "저녁"
    else:
        time_str = "심야"
    return time_str

time_str =time.map(test)
new_filltered_df["time"] = time_str
```
