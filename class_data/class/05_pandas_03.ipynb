    - 2. 전체 데이터에서 성별 데이터를 나눠서 확률 데이터로 변경
    - 3. 1~5등까지의 성별에 따른 등수 확률 데이터를 출력

```python
df_race = pd.read_csv("datas/2014_p.csv",encoding = " euc-kr") #한글파일
df_race.tail(2)
```

```python
df_horse = pd.read_csv("datas/2014_s.csv",encoding = " euc-kr") #한글파일
df_horse.tail(2)
```

```python
merge_df = pd.merge(df_horse, df_race)
merge_df.tail()
```

```python
merge_df["COUNT"] = 1
```

```python
df = merge_df.pivot_table("COUNT","RANK","SEX",aggfunc=np.sum)
df.head()
```

```python
# 1, 성별에 따른 순위 데이터 프레임으로 생성
```

```python
# 성별에 대한 전체 데이터
count = merge_df.groupby("SEX").size()
count
```

```python
# 전체 데이터에서 성별 데이터를 나눠서 확률 데이터로 변경
result = df / count*100
result.head()
```

```python
# 소수 둘째짜리까지만 출력
# 결론 : 수컷이 좋은 성적을 낼 확률이 높다
result = round(result,2)
result.head()
```

# Quiz2. 순위를 승점제로 했을 때 2014년 시즌에서 말들의 성적은?
- 1등 : 3점
- 2등 : 2점
- 3등 : 1점

```python
df = merge_df.groupby("RCHOSE_NM").agg("sum").reset_index()
```

```python
df = df.sort_values("POINT",ascending = False)[["RCHOSE_NM","POINT"]]
df.head()
```

```python
merge_df["POINT"] = merge_df["RANK"].apply(
    lambda rank: 4- rank if rank <=3 else 0
)
result = merge_df[["RCHOSE_NM","POINT"]]
result = result.sort_values("POINT",ascending = False)
result.reset_index(drop=True)
```
